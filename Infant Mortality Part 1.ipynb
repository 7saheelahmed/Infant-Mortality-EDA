```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.style.use('ggplot')
```

```python
train_df = pd.read_csv("train_infant.csv" , delimiter = ',' , header = 0)
train_df.columns = train_df.columns.str.strip()
train_df.head()
```

```python
test_df = pd.read_csv("test_infant.csv" ,  delimiter = ',' , header = 0 , skipinitialspace=True)
test_df.head()
```

```python
labels = pd.read_csv('results_infant.csv')
labels.head()
```

```python
test_df = pd.concat([test_df,labels] , axis = 1)
test_df.columns = test_df.columns.str.strip()
test_df.columns
```

```python
# renaming the first column using rename()
test_df = test_df.rename(columns = {'1':'Infant Living at Time of Report'})
test_df.columns
```

```python
frames = [train_df , test_df]
combined_df = pd.concat(frames , ignore_index = True)
combined_df.shape
```

```python
combined_df = combined_df[:-1]
combined_df.tail()
```

```python
Data_dict = pd.DataFrame(combined_df.dtypes)
Data_dict['MissingVal'] = combined_df.isnull().sum()
Data_dict['UniqueVal'] = combined_df.nunique()
Data_dict['Count'] = combined_df.count()
Data_dict = Data_dict.rename(columns = {0:'DataType'})
Data_dict
```
